```python
import numpy as np
import copy
import matplotlib.pyplot as plt
import time

import sys
sys.path.append('../MoitraRohatgi/')
sys.path.append('../')
import auditor_tools
import algorithms
import experiments
import examples
import our_experiments
```

For every instance of Boston Housing data considered by [MR'22], we first compute both their and our upper bounds (increasing the sample size to 300 to ensure run times are approximately equal). We time the run times of both methods and at the end of each block of 13 instances, we print the cumulative (combined) run time of the [MR'22] algorithms and of Gurobi (including setup times).

```python
greedy_values = []
zam_noresolve = []
our_zam = []
net_values = []
lp_values = []
MR_timer = 0
our_values = []
our_bounds = []
Gurobi_timer = 0
for a in range(13): 
    for b in range(13): 
        if a == b:
            continue
        X,y = examples.boston_housing_data([a,b])
        greedy_values.append(algorithms.sensitivity(X,y))
        start_mr=time.time()
        net_values.append(algorithms.net_algorithm(X,y,300))
        lp_values.append(algorithms.lp_algorithm_2d(X,y,[0],300))
        MR_timer+= time.time()-start_mr
        Xflipped = copy.deepcopy(np.flip(X,axis=1))
        our_zam_res = auditor_tools.ZAMinfluence_resolving(Xflipped,y)
        our_zam.append(len(X)-our_zam_res[0])
        start_g=time.time()
        our_res = auditor_tools.solve_regression_fractional(Xflipped,y, 
                                intercept=False,time_limit=30, verbose=False)
        our_values.append(len(X)-our_res[1])
        our_bounds.append(len(X)-our_res[0])
        Gurobi_timer+=time.time()-start_g
        X,y = examples.boston_housing_data([a,b])
        Xflipped = copy.deepcopy(np.flip(X,axis=1))
        zam_noresolve.append(len(X)-auditor_tools.ZAMinfluence_upper_bound(Xflipped,y)[0])
    print('MR: ', MR_timer, 'Gurobi: ', Gurobi_timer)
```

Just to verify, we check that our solutions indeed change the sign to 0.

```python
# full weights (first coefficient is negative)
algorithms.ols(X,y,np.array([1 for w in our_res[2]]))
```

```python
# weighted by gurobi solution (first coefficient is 0)
algorithms.ols(X,y,np.array([w.X for w in our_res[2]]))
```

Across all instances (on average, using the time limits we set), Gurobi runs in about 60% of the time of [MR'22].

We first compare the lower bounds identified by [MR'22] and by Gurobi by computing the fraction of isntances on which Gurobi identifies a stronge lower bounds:

```python
# Fraction of instances on which Gurobi creates a tighter lower bound
len([x for x in range(len(our_bounds)) if our_bounds[x]>lp_values[x]])/len(our_bounds)
```
